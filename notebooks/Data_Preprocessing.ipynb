```python
import pandas as pd
```

```python
# مسیر فایل دیتاست با روت استرینگ
file_path = r'C:\Users\HsH-HsH\weighted_split_criteria\notebooks\datasets\customer_churn_data.csv'

# بارگذاری داده
df = pd.read_csv(file_path)
```

```python
# بررسی تعداد مقادیر گمشده در هر ستون
print(df.isnull().sum())
```

```python
# تعداد ردیف های اولیه
initial_rows = len(df)

# حذف ردیف هایی که حداقل یک مقدار گمشده دارند
df_cleaned = df.dropna()

# تعداد ردیف های باقی مانده بعد از حذف داده های ناقص
after_dropna_rows = len(df_cleaned)

# تعداد ردیف‌های حذف شده
rows_removed = initial_rows - after_dropna_rows

print(f"تعداد ردیف‌های اولیه: {initial_rows}")
print(f"تعداد ردیف‌های بعد از حذف داده‌های ناقص: {after_dropna_rows}")
print(f"تعداد ردیف‌های حذف شده: {rows_removed}")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df_cleaned['Churn'] = le.fit_transform(df_cleaned['Churn'])
```

```python
le = LabelEncoder()
df_cleaned.loc[:, 'Churn'] = le.fit_transform(df_cleaned['Churn'])
```

```python
# نمایش نوع ستون‌ها برای شناسایی ستون‌های دستهای
print(df_cleaned.dtypes)

# انتخاب ستون‌های categorical به جز 'Churn'
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns.tolist()
if 'Churn' in categorical_cols:
    categorical_cols.remove('Churn')

# نمایش ستون‌های دستهای شناسایی شده
print("ستون‌های دستهای به جز 'Churn':", categorical_cols)
```

```python
# اجرای One-Hot Encoding روی ستون‌های دسته‌ای به جز 'Churn'
df_final = pd.get_dummies(df_cleaned, columns=['Gender', 'Payment Method'])

# نمایش چند سطر اول داده نهایی برای بررسی
print(df_final.head())

# نمایش اطلاعات کلی داده نهایی
print(df_final.info())
```

```python
# ذخیره دیتافریم به فایل CSV جدید
output_path = r'C:\Users\HsH-HsH\weighted_split_criteria\notebooks\datasets\customer_churn_data_preprocessed.csv'
df_final.to_csv(output_path, index=False)
```

```python
import os
import ast

utils_path = r'C:\Users\HsH-HsH\weighted_split_criteria\src\utils'

files = os.listdir(utils_path)
print("فایل‌های موجود در utils:", files)

def get_functions_and_classes(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        node = ast.parse(f.read())
    funcs = [n.name for n in node.body if isinstance(n, ast.FunctionDef)]
    classes = [n.name for n in node.body if isinstance(n, ast.ClassDef)]
    return funcs, classes

for file in files:
    if file.endswith('.py'):
        full_path = os.path.join(utils_path, file)
        funcs, classes = get_functions_and_classes(full_path)
        print(f"فایل: {file}")
        print(f"  توابع: {funcs}")
        print(f"  کلاس‌ها: {classes}")
        print('-' * 30)
```
