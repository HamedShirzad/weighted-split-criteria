```python
import pandas as pd

dataset_path = r'C:\Users\HsH-HsH\weighted_split_criteria\notebooks\datasets\customer_churn_data_preprocessed.csv'
df = pd.read_csv(dataset_path)

print("نمایش چند سطر اول دیتاست:")
print(df.head())

print("\nبررسی مقادیر گمشده:")
print(df.isnull().sum())

df = df.dropna()

if df['Churn'].dtype != 'int64' and df['Churn'].dtype != 'float64':
    df['Churn'] = pd.to_numeric(df['Churn'], errors='coerce')

X = df.drop(columns=['Churn'])
y = df['Churn']

X = X.apply(pd.to_numeric, errors='coerce')

mask = X.notnull().all(axis=1) & y.notnull()
X = X.loc[mask]
y = y.loc[mask]

# تبدیل مقادیر بولی به عددی
X = X.astype(int)

X_np = X.to_numpy()
y_np = y.to_numpy()

print("\nشکل آرایه‌های numpy:")
print("X_np:", X_np.shape)
print("y_np:", y_np.shape)

print("\nچند نمونه از داده‌ها:")
print("X_np نمونه:", X_np[:3])
print("y_np نمونه:", y_np[:3])
```

```python
# ================================================================
# سلول ۵: آماده‌سازی لیست معیارهای ترکیبی
# ================================================================
# در این سلول، لیستی از تاپل‌ها (نام معیار، تابع معیار) را می‌سازیم.
# این لیست به عنوان ورودی به مدل ما داده خواهد شد تا بداند از کدام
# معیارها برای ساخت درخت تصمیم استفاده کند.

try:
    all_criteria = [
       # ("gini", gini_criterion),
        ("gain_ratio", gain_ratio_criterion),
        ("twoing", twoing_criterion),
        ("normalized_gain", normalized_gain_criterion),
        ("multi_class_hellinger", multi_class_hellinger),
        ("marshall", marsh_criterion),
        ("g_statistic", g_statistic_criterion),
        ("dkm", dkm_criterion),
        ("chi_squared", chi_squared_criterion),
        ("bhattacharyya", bhattacharyya_criterion),
        ("kolmogorov_smirnov", kolmogorov_smirnov_criterion),
    ]

    print("✅ لیست معیارها با موفقیت ساخته شد.")
    print(f"   - تعداد کل معیارها: {len(all_criteria)}")
    print("این لیست آماده است تا به عنوان ورودی به مدل داده شود.")

except NameError as e:
    print(f"❌ خطا: به نظر می‌رسد توابع معیار به درستی ایمپورت نشده‌اند. {e}")
    print("لطفاً مطمئن شوید سلول ۳ (ایمپورت ماژول‌ها) به درستی اجرا شده است.")
except Exception as e:
    print(f"❌ یک خطای غیرمنتظره رخ داد: {e}")
```

```python
# ================================================================
# سلول ۶ (نسخه دیباگ - بدون try-except و با print اصلاح شده)
# ================================================================
# این نسخه از سلول، بلوک try-except را ندارد تا بتوانیم
# گزارش خطای کامل پایتون را مشاهده کنیم.

print("\n--- شروع فرآیند ساخت نمونه مدل (حالت دیباگ) ---")

# یک نمونه از مدل شما را می‌سازیم.
model = WeightedDecisionTreeModel(
    criteria_funcs_weights=all_criteria,
    max_depth=13,
    a=X_np
)

print("✅ یک نمونه از مدل WeightedDecisionTreeModel با موفقیت ساخته شد.")
# 🔥 اصلاحیه: مسیر دسترسی به max_depth اصلاح شد
print(f"   - حداکثر عمق تنظیم شده: {model.max_depth}")
print(f"   - تعداد معیارهای ترکیبی: {len(model.criteria)}")

print("\nمدل آماده برای آموزش در سلول بعدی است.")
```

```python
import numpy as np
from tabulate import tabulate
# ایمپورت‌های سفارشی شما
from model.weighted_decision_tree import SingleCriterionMetric
from custom_tree_classifier.models.decision_tree import CustomDecisionTreeClassifier

# فرض می‌کنیم متغیرهای all_criteria, X_np, y_np از قبل در حافظه هستند

print("🏁 شروع مسابقه قهرمانی معیارها برای کاهش FN (نسخه اصلاح شده)...")
print("="*50)

results = []
# (مطمئن شوید all_criteria شامل لیست تاپل‌های (نام، تابع) است)
for name, func in all_criteria:
    try:
        print(f"   - در حال تست معیار: {name}...")
        
        # ۱. ساختن متریک ساده
        metric_obj = SingleCriterionMetric(name, func)
        
        # ۲. ساختن درخت تصمیم استاندارد
        temp_model = CustomDecisionTreeClassifier(max_depth=10, metric=metric_obj)
        
        # ================================================================
        # 🔥 اصلاح کلیدی: اضافه کردن metric_data به فراخوانی fit 🔥
        # ================================================================
        # ساختن ورودی سوم مورد نیاز برای تابع fit
        metric_data = y_np.reshape(-1, 1)
        # فراخوانی fit با هر سه ورودی لازم
        temp_model.fit(X_np, y_np, metric_data)
        # ================================================================
        
        # ۴. محاسبه تعداد FN
        probas = temp_model.predict_proba(X_np)
        preds = (probas[:, 1] > 0.5).astype(int) if probas.shape[1] > 1 else np.zeros_like(y_np)
        fn_count = np.sum((y_np == 1) & (preds == 0))
        
        results.append([name, fn_count, "✅ OK"])
        
    except Exception as e:
        import traceback
        print(f"❌ ERROR in criterion {name}: {e}")
        traceback.print_exc()
        results.append([name, "N/A", "❌ ERROR"])

print("\n🏆 جدول نهایی لیگ معیارها 🏆")
# نمایش نتایج در یک جدول زیبا
print(tabulate(results, headers=["نام معیار", "تعداد FN (روی کل داده)", "وضعیت"], tablefmt="pretty"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import numpy as np

# ================================================================
# ۱. استفاده از متد predict برای دریافت پیش‌بینی‌ها با متغیرهای صحیح
# ================================================================
# از X_np برای پیش‌بینی و از y_np برای مقایسه با برچسب‌های واقعی استفاده می‌کنیم.
y_pred = model.predict(X_np)

print("--- ✅ پیش‌بینی با موفقیت انجام شد ---")

# ================================================================
# ۲. محاسبه و ترسیم ماتریس درهم‌ریختگی (Confusion Matrix)
# ================================================================
try:
    cm = confusion_matrix(y_np, y_pred)
    tn, fp, fn, tp = cm.ravel()
except ValueError:
    if np.all(y_pred == 0):
        tn = np.sum((y_np == 0)); fn = np.sum((y_np == 1)); tp, fp = 0, 0
    else:
        tp = np.sum((y_np == 1)); fp = np.sum((y_np == 0)); tn, fn = 0, 0
    cm = np.array([[tn, fp], [fn, tp]])

plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
            xticklabels=['Predicted Negative (0)', 'Predicted Positive (1)'],
            yticklabels=['Actual Negative (0)', 'Actual Positive (1)'])
ax.set_title('Confusion Matrix', fontsize=16, pad=20)
ax.set_xlabel('Predicted Label', fontsize=12)
ax.set_ylabel('True Label', fontsize=12)
plt.show()

# ================================================================
# ۳. نمایش گزارش کامل و معیارهای کلیدی
# ================================================================
print("\n" + "="*60)
print("📊 گزارش کامل ارزیابی عملکرد مدل")
print("="*60 + "\n")

accuracy = accuracy_score(y_np, y_pred)
print(f"🎯 Accuracy (دقت کل): {accuracy:.2%}")
print(f"❗ False Negatives (FN): {fn}  (تعداد نمونه‌های مثبت که به اشتباه منفی تشخیص داده شدند)")

recall_positive = tp / (tp + fn) if (tp + fn) > 0 else 0
print(f"📈 Recall for Positive Class (Class 1): {recall_positive:.2%}\n")

print("-"*(60))
print("📋 گزارش طبقه‌بندی (Classification Report):")
print("-"*(60))
print(classification_report(y_np, y_pred, target_names=['Class 0 (Negative)', 'Class 1 (Positive)'], zero_division=0))

print("\n--- ✅ تحلیل به پایان رسید ---")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
import numpy as np

# --- پیش‌بینی و آماده‌سازی داده‌ها ---
# از مدل آموزش‌دیده برای تولید پیش‌بینی‌ها و احتمالات استفاده می‌کنیم
y_pred = model.predict(X_np)
# برای نمودارهای ROC و Precision-Recall به احتمالات کلاس مثبت نیاز داریم
y_probas = model.predict_proba(X_np)[:, 1] 

print("--- ✅ پیش‌بینی‌ها برای تحلیل بصری آماده شد ---")

# --- ۱. ماتریس درهم‌ریختگی (فشرده و تمیز) ---
cm = confusion_matrix(y_np, y_pred)
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(5, 4)) # اندازه نمودار کوچکتر شده است
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=False,
            xticklabels=['Predicted Neg (0)', 'Predicted Pos (1)'],
            yticklabels=['Actual Neg (0)', 'Actual Pos (1)'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# --- ۲. منحنی ROC (Receiver Operating Characteristic) ---
fpr, tpr, _ = roc_curve(y_np, y_probas)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# --- ۳. منحنی Precision-Recall ---
precision, recall, _ = precision_recall_curve(y_np, y_probas)
avg_precision = average_precision_score(y_np, y_probas)
plt.figure(figsize=(7, 5))
plt.plot(recall, precision, color='darkcyan', lw=2, label=f'Average Precision = {avg_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="upper right")
plt.show()

# --- ۴. نمودار وزن معیارها (در گره ریشه) ---
try:
    weights_dict = model.metric.weights_dict
    criteria_names = list(weights_dict.keys())
    weights = list(weights_dict.values())
    
    plt.figure(figsize=(10, 5))
    sns.barplot(x=criteria_names, y=weights, palette='viridis')
    plt.title('Criterion Weights at Root Node')
    plt.xlabel('Criterion')
    plt.ylabel('Assigned Weight')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Could not plot weights: {e}")


# --- ۵. گزارش متنی نهایی ---
print("\n" + "="*60)
print("📊 گزارش کامل ارزیابی عملکرد مدل")
print("="*60)
print(classification_report(y_np, y_pred, target_names=['Class 0 (Negative)', 'Class 1 (Positive)'], zero_division=0))
print("\n--- ✅ تحلیل به پایان رسید ---")

```
